Func with Optional ARG

```python
def calculateEMI(amount,duration,downPayment=0):
    loan_amount = amount - downPayment
    emi=loan_amount / duration
    return emi
calculateEMI(120000, 12)
```

try except

```python
def divide(a:int,b:int)->None:
    try:
        result=a/b
    except Exception as e:
        print("Error:", e)
    else:
        print("Result:", result)
    finally:
        print("Execution completed.")
divide(10,0)
divide(10,2)
```

Docstrings for Documentation

```python
def divide(a:int,b:int)->None:
    """Divides two numbers.

       Args:
           a (int): The numerator.
           b (int): The denominator.

       Returns:
           float: The result of the division or None if an error occurs.
    """
    try:
        result=a/b
    except Exception as e:
        print("Error:", e)
    else:
        print("Result:", result)
    finally:
        print("Execution completed.")

#documentation
help(divide)

#test cases
divide(10,0)
divide(10,2)
```
